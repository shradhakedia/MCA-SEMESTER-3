```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,cross_val_score, cross_val_predict
from sklearn import metrics
```

```python
df = pd.read_csv('abalone.csv')

df['Male'] = df['Sex']
df['Female'] = df['Sex']

df.drop(["Sex"],axis=1, inplace=True)

df['Male'] = df['Male'].map({'M' : 1, 'F' : 0, 'I' : 0})
df['Female'] = df['Female'].map({'M' : 0, 'F' : 1, 'I' : 0})
```

```python
y = df.iloc[:, 7]

df.drop(["Rings"],axis=1, inplace=True)

x = df.iloc[:,:]

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

# 1. For Test set = 33.33%

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3333)
```

## i) Naive Bayes

```python
gaussian_nb = GaussianNB()
y_prediction = gaussian_nb.fit(x_train, y_train).predict(x_test)
print("Accuracy is :", metrics.accuracy_score(y_test, y_prediction)*100, "%")
```

## ii) K-NN

```python
k_nn = KNeighborsClassifier()
y_prediction_knn= k_nn.fit(x_train,y_train).predict(x_test)
print("Accuracy is : ", metrics.accuracy_score(y_test,y_prediction_knn)*100, "%")
```

## iii) Decision Tree

```python
DT_classifier= DecisionTreeClassifier(random_state=1)
y_prediction_DT_classifier= DT_classifier.fit(x_train, y_train).predict(x_test)
print("Accuracy is :", metrics.accuracy_score(y_test, y_prediction_DT_classifier)*100, "%")
```
